```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import OrderedDict
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
def singletrace(path, plottype, label, color, smooth=False):
    
    dfdat = pd.read_csv(path, delimiter='\t', index_col=0)
    dat = dfdat.reset_index().values
    x = dat[:,0]
    y = np.sqrt(dat[:,1])
    
    if smooth == True:
        y = lowess(y, x, is_sorted=True, frac=0.001, it=0)[:,1]
        label = label + ' smthd'
            
    if plottype == 'loglog':
        plt.loglog(x, y, c=color, alpha=0.7, lw=2, label=label)
    elif plottype == 'semilogy':
        plt.semilogy(x, y, c=color, alpha=0.7, lw=2, label=label)
            
    plt.xlabel('Frequency / Hz')
    plt.ylabel('Noise / $(m/s)_{rms} Hz^{-0.5}$')
    plt.ylim(2e-10,2e-6)
    plt.xlim(0.5,100)
    plt.grid(b=1, which='major', axis='both', lw=1)
    plt.grid(b=1, which='minor', axis='both', lw=0.3)
    plt.legend()
    return
```

```python
pump_stab = r'G:\Accelerometer\Logs\Laboratory Logs\CRYO-retest20190822\pump_stab\20190822_135927-10kHz-100s-10reps.txt-AVG.txt'
nopump_stab = r'G:\Accelerometer\Logs\Laboratory Logs\CRYO-retest20190822\nopump_stab\20190822_143915-10kHz-100s-10reps.txt-AVG.txt'
floor_nopump = r'G:\Accelerometer\Logs\Laboratory Logs\CRYO-retest20190822\floor_nopump\20190822_164616-10kHz-100s-10reps.txt-AVG.txt'

singletrace(pump_stab, 'loglog', 'cryo_pump_stab', 'b', smooth=True)
singletrace(nopump_stab, 'loglog', 'cryo_nopump_stab', 'r', smooth=True)
singletrace(floor_nopump, 'loglog', 'floor_nopump', 'gold', smooth=True)

plt.title('cryo-20190822')
plt.show()
```
